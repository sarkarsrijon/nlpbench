```python
import pandas as pd

# Load files
my_df = pd.read_excel(r"D:\emory books\2026 spring\qtm 329\claude_lemmas.xlsx")             # Generated lemma file (Word, Lemma columns)
user_df = pd.read_excel(r"D:\emory books\2026 spring\qtm 329\Unimorph.xlsx", header=None, names=['user_lemma', 'Word', 'tags'])

# Merge on matching words
merged = user_df.merge(my_df, on='Word', how='inner')

# Add match column: 1 if lemma matches, 0 if not
merged['match'] = (merged['user_lemma'].str.strip().str.lower() == 
                   merged['Lemma'].str.strip().str.lower()).astype(int)

# Rename for clarity
merged = merged[['Word', 'user_lemma', 'Lemma', 'match']]
merged.columns = ['Word', 'Unimorph_lemma', 'generated_lemma', 'match']

# Save output
merged.to_excel(r'D:\emory books\2026 spring\qtm 329\comparison_output.xlsx', index=False)
print(merged.head(20))
print(f"\nTotal matches: {merged['match'].sum()} / {len(merged)}")
```
