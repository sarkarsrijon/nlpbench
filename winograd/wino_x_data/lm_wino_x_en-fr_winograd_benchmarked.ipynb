```python
import pandas as pd
import json

entries = []
with open(r"D:\emory books\2026 spring\qtm 329\winogrande_1.1\train_l.jsonl", 'r') as f:
    for line in f:
        entries.append(json.loads(line.strip()))

df = pd.DataFrame(entries)

df.to_excel(r'D:\emory books\2026 spring\qtm 329\parsed_entries.xlsx', index=False)
print(df.head())
print(f"\nTotal entries: {len(df)}")
```

```python
import pandas as pd

df = pd.read_excel(r'D:\emory books\2026 spring\qtm 329\winograd\wino_x_data\lm_wino_x\lm_wino_x_en-fr_only.xlsx')

# Replace answer number with the actual word
def get_answer_word(row):
    if str(row['answer']) == '1':
        return row['option1_fr']
    elif str(row['answer']) == '2':
        return row['option2_fr']
    return row['answer']

df['answer_word'] = df.apply(get_answer_word, axis=1)

df.to_excel(r"D:\emory books\2026 spring\qtm 329\winograd\wino_x_data\lm_wino_x\lm_wino_x_en-fr_only_with_words.xlsx", index=False)
print(df[['option1_fr', 'option2_fr', 'answer', 'answer_word']].head(10))
```

```python
import pandas as pd

df = pd.read_excel(r"D:\emory books\2026 spring\qtm 329\winograd\wino_x_data\lm_wino_x\lm_wino_x_en-fr_only_with_words.xlsx")

df['match'] = (df['answer_claude'].str.strip().str.lower() == 
               df['answer_word'].str.strip().str.lower()).astype(int)

df.to_excel(r"D:\emory books\2026 spring\qtm 329\winograd\wino_x_data\lm_wino_x\lm_wino_x_en-fr_only_with_match.xlsx", index=False)
print(df[['answer_claude', 'answer_word', 'match']].head(10))
print(f"\nTotal matches: {df['match'].sum()} / {len(df)}")
```
