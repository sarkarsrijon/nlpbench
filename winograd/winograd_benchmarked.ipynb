```python
import pandas as pd
import json

entries = []
with open(r"D:\emory books\2026 spring\qtm 329\winogrande_1.1\train_l.jsonl", 'r') as f:
    for line in f:
        entries.append(json.loads(line.strip()))

df = pd.DataFrame(entries)

df.to_excel(r'D:\emory books\2026 spring\qtm 329\parsed_entries.xlsx', index=False)
print(df.head())
print(f"\nTotal entries: {len(df)}")
```

```python
import pandas as pd

df = pd.read_excel(r'D:\emory books\2026 spring\qtm 329\parsed_entries.xlsx')

# Replace answer number with the actual word
def get_answer_word(row):
    if str(row['answer']) == '1':
        return row['option1']
    elif str(row['answer']) == '2':
        return row['option2']
    return row['answer']

df['answer_word'] = df.apply(get_answer_word, axis=1)

df.to_excel(r"D:\emory books\2026 spring\qtm 329\parsed_entries_with_words.xlsx", index=False)
print(df[['option1', 'option2', 'answer', 'answer_word']].head(10))
```

```python
import pandas as pd

df = pd.read_excel(r"D:\emory books\2026 spring\qtm 329\parsed_entries_with_words.xlsx")

df['match'] = (df['filled_word'].str.strip().str.lower() == 
               df['answer_word'].str.strip().str.lower()).astype(int)

df.to_excel(r"D:\emory books\2026 spring\qtm 329\parsed_entries_with_match.xlsx", index=False)
print(df[['filled_word', 'answer_word', 'match']].head(10))
print(f"\nTotal matches: {df['match'].sum()} / {len(df)}")
```
